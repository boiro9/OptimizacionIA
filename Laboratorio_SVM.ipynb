## Implementaci贸n de SVM en Pyomo

**Descipci贸n de SVM**  

Supongamos que tenemos $l$ observaciones $\{\pmb{x}^{1},\ldots,\pmb{x}^{l}\}\subset \mathbb{R}^{n}$. Para cada observaci贸n $i\in \{1,\ldots,l\}$ tenemos:  
 - Los valores de $n$ variables explicativas $x_{1}^{i},\ldots,x_{n}^{i}$.
 - Su clasificaci贸n $y_{i}\in \{-1,1\}$.

El problema primal para llevar a cabo una **clasificaci贸n mediante SVMs** ser铆a el siguiente:
$$
\begin{array}{rl}
\min       & \frac{1}{2}\pmb{w}^{T}\pmb{w}=\frac{1}{2}\sum_{j=1}^{n}(w_{j})^{2} \\
\text{s.a.}& y_{i}(\pmb{w}^{T}\pmb{x}^{i}+b)-1 \geq 0, \, i=1,\ldots,l \Leftrightarrow y_{i}(\sum_{j=1}^{n}w_{j}x_{j}^{i}+b)-1\geq 0, \, i=1,\ldots,l \\
		   & \pmb{w}\in \mathbb{R}^{n}, b\in \mathbb{R}
\end{array}
$$
